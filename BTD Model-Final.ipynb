# BRAIN TUMOR PREDICTION USING CNN

## THIS FILE HAS MODEL BUILDING


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Convolution2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen =  ImageDataGenerator(rescale = 1./255,shear_range = 0.2,zoom_range = 0.2,horizontal_flip = True)
test_datagen =  ImageDataGenerator(rescale = 1./255)
```

```python
x_train = train_datagen.flow_from_directory("BTD/train",target_size = (64,64),batch_size = 32, class_mode ="categorical")
x_test = test_datagen.flow_from_directory("BTD/test",target_size = (64,64),batch_size = 32, class_mode ="categorical")
# x_train = train_datagen.flow_from_directory("BTD/train",target_size = (64,64),batch_size = 8, class_mode ="categorical")
# x_test = test_datagen.flow_from_directory("BTD/test",target_size = (64,64),batch_size = 8, class_mode ="categorical")
```

```python
model = Sequential()
model.add(Convolution2D(32,(3,3), input_shape=(64,64,3)))
model.add(MaxPooling2D(2,2))
model.add(Flatten())
```

```python
model.add(Dense(units = 256,activation = "relu",kernel_initializer = "random_uniform"))
model.add(Dense(units = 256,activation = "relu",kernel_initializer = "random_uniform"))
model.add(Dense(units = 2,activation = "softmax",kernel_initializer = "random_uniform"))
model.compile("sgd",loss = "categorical_crossentropy",metrics = ["accuracy"])
```

```python
model.fit_generator(x_train ,steps_per_epoch = 7 ,epochs = 100,validation_data= x_test , validation_steps = 1)
# model.fit_generator(x_train ,steps_per_epoch = 24 ,epochs = 100,validation_data= x_test , validation_steps = 8)
```

```python
model.save("BTD2.h5")
```
